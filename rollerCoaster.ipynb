Call your function with the roller coaster DataFrame and an amusement park name.

```python
from numpy.random import seed
from numpy.random import shuffle
from numpy.random import randint
seed()

def park_bar(coasters, park_name):
    # Select the rows for the parkname
    x = coasters[coasters['park'] == park_name]
    # Remove the row that are NaN
    x = x.fillna(0)
    xpos= range(len(x['name']))
    plt.bar(xpos, x['num_inversions'])
    plt.title(park_name + ' - Number of Inversions' )

park_names = list(set(coasters.park))
shuffle(park_names)

plt.figure(figsize=(8,8))
plt.subplot(2, 1, 1)
park_bar(coasters, park_names[0])
plt.subplot(2, 1, 2)
park_bar(coasters, park_names[1])
plt.show()
```

Write a function that creates a pie chart that compares the number of operating roller coasters ('status.operating') to the number of closed roller coasters ('status.closed.definitely'). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

Call your function with the roller coaster DataFrame.

```python
def operating(coasters):
    
    # Count number of unique_key for each status  
    x = coasters.groupby('status').count().reset_index()
    
    # Select rows with matching status
    x = x[x.status.isin(['status.operating', 'status.closed.definitely'])].reset_index()

    # Make the figure wide enough for the legend
    fig = plt.figure(figsize=(8,5))
    
    # Set aspect ratio=1 so pie is circular
    plt.axis('equal')
    
    # Any column works
    plt.pie(x.name, labels=list(x.status), autopct="%0.1f%%")
    
    # Add a legend even though labels are present
    plt.legend(x.status, loc='lower right')
    plt.title("Operational Rollercoasters")
    plt.show()
    
###############################################

operating(coasters)

```

.scatter() is another useful function in matplotlib that you might not have seen before. .scatter() produces a scatter plot, which is similar to .plot() in that it plots points on a figure. .scatter(), however, does not connect the points with a line. This allows you to analyze the relationship between to variables. Find .scatter()‘s documentation https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html
